```python
# temp_sen = u""
# 한글 깨지기 때문에 pprint 함수 사용하여 한글 출력.
cutdata = {}

for prop in props:
    temp_sen = prop
    cut = twi_tagger.nouns(temp_sen)
    # pprint(cut)
    # print(cut)
    cutdata[temp_sen] = cut
```

```python
def calculateSim(propi, propj):

    values = []
    value1 = distance.jaccard(propi, propj)
    value2 = distance.jaccard(propi, propj)
    value3 = distance.jaccard(propi, propj)
    values.append(value1)
    values.append(value2)
    values.append(value3)
    return values
```

```python
print(calculateSim('출생지', '사망지'))
```

```python
for i in range(10):
    props1 = random.choice(props)
    props2 = random.choice(props)
    val = calculateSim(props1, props2)
    print(props1, props2)
    print(val)  # Equi, Subsump, Opposite
```
